```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
```

```python
df = pd.read_csv("Most Streamed Spotify Songs 2024.csv", encoding='latin1')
```

```python
df.head()
```

```python
# Standardize column names: lowercase and replace spaces with underscores
df.columns = df.columns.str.lower().str.replace(' ', '_').str.strip()

# Print the updated column names
print(df.columns)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
#changing Dtype
df['all_time_rank'] = df['all_time_rank'].str.replace(',', '').astype('int64')
df['spotify_streams'] = df['spotify_streams'].astype(str).str.replace(',', '').astype('float64')

```

```python
# List of columns to convert
columns_to_convert = [
    'spotify_playlist_count', 'spotify_playlist_reach', 'youtube_views', 'youtube_likes', 
    'tiktok_posts', 'tiktok_likes', 'tiktok_views', 'youtube_playlist_reach'
]

# Loop through each column and convert
for column in columns_to_convert:
    df[column] = pd.to_numeric(df[column].astype(str).str.replace(',', ''), errors='coerce')

# Check the results
df.info()
```

```python
#new and updated statical description
df.describe()
```

```python
df.isnull().sum()
```

Handling missing data

```python
df[df['artist'].isnull()]
```

```python
df['artist'].fillna('Unknown', inplace=True)
```

```python
mean_value = df['spotify_streams'].mean()
df['spotify_streams'].fillna(mean_value, inplace=True)
```

```python
mean_playlist_count = df['spotify_playlist_count'].mean()
df['spotify_playlist_count'].fillna(mean_playlist_count, inplace=True)

mean_playlist_reach = df['spotify_playlist_reach'].mean()
df['spotify_playlist_reach'].fillna(mean_playlist_reach, inplace=True)
```

```python
# Replace missing values with median 
df['spotify_popularity'].fillna(df['spotify_popularity'].median(), inplace=True)
df['youtube_views'].fillna(df['youtube_views'].median(), inplace=True)
df['youtube_likes'].fillna(df['youtube_likes'].median(), inplace=True)
df['tiktok_posts'].fillna(df['tiktok_posts'].median(), inplace=True)
df['tiktok_likes'].fillna(df['tiktok_likes'].median(), inplace=True)
df['tiktok_views'].fillna(df['tiktok_views'].median(), inplace=True)
df['youtube_playlist_reach'].fillna(df['youtube_playlist_reach'].median(), inplace=True)
df['apple_music_playlist_count'].fillna(df['apple_music_playlist_count'].median(), inplace=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns for plotting
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Set up the figure and axis for the plots
plt.figure(figsize=(15, 15))

# Plot histograms for each numerical feature
for i, col in enumerate(numerical_columns):
    plt.subplot(5, 5, i + 1)  # Arrange in a 5x5 grid
    sns.histplot(df[col], kde=True, bins=30, color='skyblue')
    plt.title(f'Distribution of {col}')
    plt.tight_layout()

plt.show()
```

```python
# Ensure all columns are float64
df[numerical_columns] = df[numerical_columns].apply(pd.to_numeric, errors='coerce')

# Try plotting one boxplot to debug
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['spotify_streams'], color='lightgreen')
plt.title('Boxplot of Spotify Streams')
plt.show()
```

```python
# Calculate the Q1 (25th percentile) and Q3 (75th percentile) for 'spotify_streams'
Q1 = df['spotify_streams'].quantile(0.25)
Q3 = df['spotify_streams'].quantile(0.75)
IQR = Q3 - Q1

# Define the outlier boundaries
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df[(df['spotify_streams'] < lower_bound) | (df['spotify_streams'] > upper_bound)]

# Print the outliers
print("Outliers in spotify_streams:\n", outliers)

# Handle outliers (example: remove them or replace with the median)
df['spotify_streams'] = df['spotify_streams'].apply(
    lambda x: df['spotify_streams'].median() if x < lower_bound or x > upper_bound else x
)

# After handling outliers, let's plot the boxplot again to see the effect
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['spotify_streams'], color='lightgreen')
plt.title('Boxplot of Spotify Streams (After Outlier Handling)')
plt.show()
```
